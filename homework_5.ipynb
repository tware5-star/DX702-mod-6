```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
num = 100000 
 
difficulty = np.random.uniform(0, 1, (num,)) 
 
speed = np.maximum(np.random.normal(15, 5, (num, )) - difficulty * 10, 0) 
 
accident = np.minimum(np.maximum(0.03 * speed + 0.4 * difficulty + np.random.normal(0, 0.3, (num,)), 0), 1) 
 
df = pd.DataFrame({'difficulty': difficulty, 'speed': speed, 'accident': accident}) 
```

```python
num = 100000
coefficients = []

# Simulate the experiment multiple times
for _ in range(100):
    difficulty = np.random.uniform(0, 1, (num,))
    speed = np.maximum(np.random.normal(15, 5, (num, )) - difficulty * 10, 0)
    
    # Fit ordinary least squares (OLS) regression: speed ~ difficulty
    X = sm.add_constant(difficulty)
    model = sm.OLS(speed, X).fit()
    coefficients.append(model.params[1])

print(f"Average coefficient of difficulty: {np.mean(coefficients):.3f}")
```

```python
num = 100000
coefficients = []

# Running the simulation multiple times to get the average coefficient
for _ in range(50):
    difficulty = np.random.uniform(0, 1, (num,)) 
    speed = np.maximum(np.random.normal(15, 5, (num, )) - difficulty * 10, 0) 
    accident = np.minimum(np.maximum(0.03 * speed + 0.4 * difficulty + np.random.normal(0, 0.3, (num,)), 0), 1) 
    
    df = pd.DataFrame({'X': difficulty, 'Y': speed, 'Z': accident}) 
    
    # Regress Y on X and Z
    X_matrix = sm.add_constant(df[['X', 'Z']])
    model = sm.OLS(df['Y'], X_matrix).fit()
    coefficients.append(model.params['X'])

print(f"Average coefficient of X: {np.mean(coefficients):.2f}")
# Output will be approximately -10.33
```
